## Module Imports

The Pyplot package is a sub-package of the matplotlib package. 

To help demonstrate the Pyplot package, the Numpy package is imported. One use of this package is to generate random data that can be used in plots.

```python
# import libraries needed 
import numpy as np
import matplotlib.pyplot as plt
```

## Simple Plot
The .arange and .sin functions in Numpy return lists of data. The .plot function plots the numbers in these lists on a two dimensional graph. 

```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)
```

## Polar Plot
Below is an example of a Polar Plot. Polar coordinates are used in mathematics as an alternative to cartesian coordinates. 

```python
# r will be the circle radius
# In this case, it is a range between

r = np.arange(0, 2, 0.01)
# theta is the circle circumference
theta = 2 * np.pi * r

plt.polar(theta, r)
```

## Box  Plot
Box plots show the spread of data across its four quartiles. Outliers are shown using the "whiskers". 

```python
# A list of 500 random numbers with a mean of 0 and standard deviation of 30 is generated to demonstrate the boxplot. 
z = np.random.normal(loc=0, size=500, scale=30)
plt.boxplot(z)
```

## Violin  Plot
Violin plots are a variation on the boxplot, showing an estimate of the density probability function

```python
# Plots the same data as that used for the boxplot previously. 
plt.violinplot(z)
```
